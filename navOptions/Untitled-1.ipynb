```python
import matplotlib.pyplot as plt

# Data from the banking Excel file
financial_factors = [
    "Credit Risk Analysis", 
    "Investment Planning", 
    "Loan Interest Calculation", 
    "Liquidity Assessment", 
    "Portfolio Diversification", 
    "Market Trend Prediction"
]

# Impact scores for plotting
risk_management_scores = [40, 25, 30, 20, 35, 15]
financial_performance_scores = [25, 35, 20, 15, 40, 30]

# Create a bar plot
x = range(len(financial_factors))  # Create a range of numbers for the x-axis

plt.figure(figsize=(12, 8))

# Plotting two sets of bars for comparison
plt.bar(x, risk_management_scores, width=0.4, label='Risk Management Impact', align='center', color='skyblue')
plt.bar([i + 0.4 for i in x], financial_performance_scores, width=0.4, label='Financial Performance Impact', color='orange')

# Add labels and title
plt.xlabel('Financial Factors', fontsize=12)
plt.ylabel('Impact Score (%)', fontsize=12)
plt.title('Impact of Financial Factors in Banking', fontsize=14)
plt.xticks([i + 0.2 for i in x], financial_factors, rotation=45, ha='right')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()







```
